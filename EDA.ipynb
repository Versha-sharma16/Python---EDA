#Question 1 : What is the difference between AI, ML, DL, and Data Science? Provide a brief explanation of each. (Hint: Compare their scope, techniques, and applications for each.)


Ans - Here’s a brief breakdown of each concept, highlighting their differences in scope, techniques, and applications:

##1. Artificial Intelligence (AI):


- Scope: AI is the broadest field and focuses on creating systems that can perform tasks that would normally require human intelligence, such as reasoning, problem-solving, decision-making, perception, and language understanding.

- Techniques: Includes symbolic AI, rule-based systems, expert systems, search algorithms, and knowledge representation, among others.

- Applications: AI is used in various domains such as robotics, game playing, voice assistants (like Siri), self-driving cars, recommendation systems, and more.

##2. Machine Learning (ML):

- Scope: ML is a subset of AI that focuses specifically on building algorithms that allow systems to learn from data and improve performance over time without being explicitly programmed.

- Techniques: Includes supervised learning, unsupervised learning, reinforcement learning, and semi-supervised learning, along with specific algorithms like decision trees, support vector machines (SVM), k-means clustering, and random forests.

- Applications: ML is widely used in areas such as spam filtering, predictive analytics, image and speech recognition, fraud detection, customer segmentation, and recommendation systems.

##3. Deep Learning (DL):

- Scope: DL is a subset of ML and is primarily concerned with neural networks that are deep (having many layers). It focuses on automatic feature extraction and learning complex patterns from large datasets.

- Techniques: DL uses deep neural networks, including convolutional neural networks (CNNs), recurrent neural networks (RNNs), and transformers. The key to deep learning is its ability to learn from vast amounts of data without much manual feature engineering.

- Applications: DL excels in tasks such as image recognition (e.g., facial recognition), natural language processing (e.g., language translation, text generation), speech recognition, and autonomous driving.

##4. Data Science:

- Scope: Data Science is a multidisciplinary field that encompasses methods from statistics, mathematics, computer science, and domain-specific knowledge to extract insights and make decisions based on data. It involves the entire process of data collection, cleaning, exploration, modeling, and interpretation.

- Techniques: Data science techniques include data visualization, statistical analysis, data mining, machine learning, and predictive modeling. It also involves tools and languages like Python, R, SQL, and data analysis libraries (e.g., pandas, NumPy, Scikit-learn).

- Applications: Data science is applied in diverse fields such as healthcare (predictive analytics), finance (risk modeling), marketing (customer segmentation), sports analytics, and business intelligence.


###Summary of key differences

- AI vs. ML: AI is the comprehensive goal of creating intelligent machines, while ML is a method for achieving AI by training a system to learn from data. All ML is considered AI, but not all AI relies on ML.

- ML vs. DL: DL is a more advanced and powerful form of ML. The primary difference is how they process data; DL uses complex, multi-layered neural networks to automatically learn features from large, unstructured datasets, while traditional ML often requires manual feature extraction and can operate on smaller datasets.

- Data Science vs. AI/ML: Data Science is the overarching process of using various techniques to extract meaningful insights from data, with the goal of making informed decisions. It is the "science of data" and uses AI and ML as tools within its toolkit to build predictive models and solve complex problems.

#Question 2: Explain overfitting and underfitting in ML. How can you detect and prevent them ? Hint: Discuss bias-variance tradeoff, cross-validation, and regularization techniques.


Ans - Overfitting and Underfitting in Machine Learning

##1. Overfitting:

 Overfitting occurs when a machine learning model learns not just the underlying patterns in the training data, but also the noise and random fluctuations. This leads to the model performing exceptionally well on training data but poorly on unseen test data because it has essentially "memorized" the training data instead of learning generalizable patterns.

1. Symptoms:

- High accuracy on training data but low accuracy on test data.

- The model has become too complex, capturing irrelevant details (noise) that do not generalize to other datasets.

2. Why it happens: Overfitting occurs when the model is too complex relative to the amount of training data or when it has too many parameters. This happens when the model tries to fit every tiny variation in the data.

##2. Underfitting:

Underfitting happens when the model is too simple to capture the underlying patterns in the data, resulting in poor performance on both the training and test datasets. The model fails to learn important relationships and cannot make accurate predictions.

1. Symptoms:

- Low accuracy on both training and test data.

- The model is not complex enough to capture the patterns in the data.

2. Why it happens: Underfitting occurs when the model is too simplistic (e.g., too few features, too simple an algorithm, or not enough training) or when it lacks enough capacity to learn complex patterns in the data.

##Bias-Variance Tradeoff:

The concepts of bias and variance are essential to understanding overfitting and underfitting, and they are at the core of the bias-variance tradeoff.

- Bias refers to errors made by the model due to overly simplistic assumptions, which can lead to underfitting. A high-bias model may miss important trends in the data.

- Variance refers to errors made by the model due to sensitivity to small fluctuations in the training data, which can lead to overfitting. A high-variance model might capture noise in the data as if it were a pattern.

- Bias-Variance Tradeoff:

- High bias leads to underfitting, where the model is too simplistic and cannot capture the underlying patterns.

- High variance leads to overfitting, where the model is too complex and captures even the noise in the data.

 -The goal is to strike a balance between bias and variance to build a model that is neither too simple nor too complex.

##How to Detect Overfitting and Underfitting:

- Overfitting:

High performance on the training data but poor performance on the test data.

The model has high variance but low bias.

- Underfitting:

Poor performance on both training and test data.

The model has high bias but low variance.

##How to Prevent Overfitting and Underfitting:

1. Cross-Validation:

- What it is: Cross-validation is a technique used to assess the performance of a model by dividing the data into multiple subsets or "folds." The model is trained on some folds and tested on the remaining folds. This process is repeated several times, and the model's performance is averaged to give a more reliable estimate of how well it generalizes to unseen data.

- How it helps: Cross-validation helps prevent overfitting by ensuring the model is tested on multiple subsets of the data, which provides a better understanding of how it will perform on unseen data. It also helps in detecting underfitting if the model consistently performs poorly across different folds.

2. Regularization Techniques:

- What it is: Regularization is a method to reduce overfitting by adding a penalty to the model's complexity. It prevents the model from fitting the noise in the training data by discouraging overly large weights or overly complex models.

###Common Regularization Methods:

- L2 Regularization (Ridge): Adds the sum of the squared values of the weights as a penalty to the loss function. This helps in preventing the model from relying too much on any one feature.

- L1 Regularization (Lasso): Adds the sum of the absolute values of the weights as a penalty. Lasso regularization can also perform feature selection by driving some weights to zero.

- Elastic Net: A combination of L1 and L2 regularization that balances the benefits of both.

- How it helps: Regularization limits the model’s complexity, helping to avoid overfitting by reducing the likelihood that it will memorize noise or irrelevant details in the training data.

##3. Model Complexity:

Simplifying the Model: If your model is overfitting, you might reduce its complexity by using simpler algorithms (e.g., switching from a deep neural network to a decision tree) or reducing the number of features (via feature selection).

Adding More Data: Increasing the amount of training data can help the model generalize better and avoid overfitting, especially for more complex models like deep learning models.

##4. Early Stopping (for iterative models like Neural Networks):

What it is: In deep learning or other iterative models, early stopping involves halting the training process before the model becomes too complex and starts overfitting to the training data. The training process is stopped when the performance on the validation data stops improving.

How it helps: Early stopping helps to prevent overfitting by ensuring the model doesn't train too long, which could lead it to start fitting noise in the data.

#Question 3:How would you handle missing values in a dataset? Explain at least three methods with example. Hint: Consider deletion, mean/median imputation, and predictive modeling.


Ans - Missing values are common in real-world datasets and can lead to incorrect analysis or model training if not handled properly. The method you choose to handle missing data depends on the nature of the data, the amount of missingness, and the impact on the final outcome.

1. Deletion (Listwise or Pairwise Deletion)
###What it is:

- Listwise Deletion: Remove entire rows that contain any missing value.

- Pairwise Deletion: Use only available data in each analysis or computation, without deleting the whole row.

###When to use:

When the number of missing entries is small

When data is missing completely at random (MCAR).

##Example:
```
import pandas as pd

df = pd.DataFrame({
    'Age': [25, 30, None, 45],
    'Salary': [50000, None, 60000, 80000]
})

# Drop rows with any missing values
df_cleaned = df.dropna()
```
###Pros:

Simple and fast.

Doesn't introduce bias if data is MCAR.

###Cons:

Risk of losing important data, especially if many rows are dropped.

2. Imputation (Mean/Median/Mode)
###What it is:

Replace missing values with a statistical estimate:

- Mean: For normally distributed numerical data.

- Median: For skewed numerical data.

- Mode: For categorical data.

###When to use:

When missing data is not extensive.

When the data is assumed to be missing at random.

##Example:
```
# Impute missing values in 'Age' with the mean
df['Age'] = df['Age'].fillna(df['Age'].mean())

# Impute missing 'Salary' with the median
df['Salary'] = df['Salary'].fillna(df['Salary'].median())
```
###Pros:

Easy and fast to implement.

Preserves dataset size.

###Cons:

Can reduce variability.

May introduce bias if the distribution is not considered.

3. Predictive Modeling (Using ML to Predict Missing Values)
###What it is:

Use other features in the dataset to predict and fill in missing values using regression, classification, or KNN imputation.

###When to use:

When missing data is significant or not MCAR.

When relationships between features can help estimate missing values.

##Example (Using KNN Imputer):

```from sklearn.impute import KNNImputer
import numpy as np

df = pd.DataFrame({
    'Age': [25, 30, np.nan, 45],
    'Salary': [50000, np.nan, 60000, 80000]
})

imputer = KNNImputer(n_neighbors=2)
df_imputed = pd.DataFrame(imputer.fit_transform(df), columns=df.columns)
```
###Pros:

More accurate than simple imputation.

Utilizes relationships in the data.

###Cons:

More computationally expensive.

Risk of overfitting or incorrect predictions if the model is not well-tuned.

#Question 4:What is an imbalanced dataset? Describe two techniques to handle it(theoritical + practical). Hint: Discuss SMOTE, Random Under/Oversampling, and class weights in models.


Ans - An imbalanced dataset is one where the distribution of classes in the target variable is uneven — i.e., one class significantly outnumbers the others.

##Example:

In a binary classification problem:

- Class 0: 950 samples

- Class 1: 50 samples

Here, the model may predict Class 0 most of the time and still achieve 95% accuracy, but fail to learn meaningful patterns for Class 1 (the minority class). This is problematic, especially in critical applications like fraud detection, disease diagnosis, or spam detection.

Problems Caused by Imbalanced Datasets:

- Models are biased toward the majority class.

- Poor recall and precision for the minority class.

- Metrics like accuracy become misleading.

###Techniques to Handle Imbalanced Datasets
1. Random Under-Sampling and Over-Sampling


Random Under-Sampling (RUS)

- Theory: Reduces the number of majority class samples by randomly removing them to balance the dataset.

- Use when: Dataset is large; losing some majority samples won't hurt model performance.

###Pros:

Simple and quick.

Reduces training time.

###Cons:

Risk of losing important data.

Can lead to underfitting.

Example:
```
from imblearn.under_sampling import RandomUnderSampler
from collections import Counter

X_resampled, y_resampled = RandomUnderSampler().fit_resample(X, y)
print(Counter(y_resampled))
```
Random Over-Sampling (ROS)

- Theory: Duplicates random samples from the minority class to balance the dataset.

- Use when: Minority class is too small and under-represented.

###Pros:

Simple to implement.

Retains all original data.

###Cons:

Can lead to overfitting due to duplication.

Example:
```
from imblearn.over_sampling import RandomOverSampler

X_resampled, y_resampled = RandomOverSampler().fit_resample(X, y)
print(Counter(y_resampled))
```

2. SMOTE (Synthetic Minority Oversampling Technique)
Theory:

SMOTE creates synthetic samples for the minority class rather than just duplicating existing ones. It selects a sample from the minority class and generates new, similar samples by interpolating between it and its nearest neighbors.

###Pros:

Reduces overfitting compared to random oversampling.

Produces more diverse synthetic samples.

###Cons:

May create ambiguous samples near class boundaries.

Can be computationally intensive.

Example:
```
from imblearn.over_sampling import SMOTE

smote = SMOTE()
X_resampled, y_resampled = smote.fit_resample(X, y)
print(Counter(y_resampled))
```
3. Class Weights in Models
Theory:

Some models (like logistic regression, SVM, decision trees) allow you to assign higher weights to minority classes during training. This helps the model pay more attention to misclassifications of minority samples.

###Pros:

No change to the dataset.

Works well with tree-based and linear models.

###Cons:

Requires careful tuning.

May not fully solve imbalance in extreme cases.

Example (with Scikit-learn’s LogisticRegression):
```
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(class_weight='balanced')
model.fit(X_train, y_train)
```

Or you can provide custom weights:
```
model = LogisticRegression(class_weight={0:1, 1:5})
```

#Question 5: Why is feature scaling important in ML? Compare Min-Max scaling and standardization. Hint: Explain impact on distance-based algorithms (e.g., KNN, SVM) and gradient descent.

Ans - What is Feature Scaling?

Feature scaling is the process of bringing all features (variables) to a similar scale — so that no single feature dominates due to its larger magnitude.

##Why It’s Important:

Many machine learning algorithms are sensitive to the scale of features. If one feature ranges from 1–1000 (e.g., income) and another from 0–1 (e.g., probability), the model may give more weight to the larger feature, even if it's less important.

###Key Areas Affected by Feature Scaling:
##1. Distance-Based Algorithms:

Algorithms like K-Nearest Neighbors (KNN), Support Vector Machines (SVM), and K-Means rely on distance metrics (e.g., Euclidean distance).

If features are not scaled, large-magnitude features dominate the distance calculation.

###Example:
```
Distance = sqrt((x1 - y1)² + (x2 - y2)²)

```
If x1 ranges from 1–1000 and x2 from 0–1, the contribution of x2 becomes negligible unless scaling is applied.

##2. Gradient Descent-Based Algorithms:

Algorithms like Logistic Regression, Neural Networks, and Linear Regression use gradient descent to minimize loss.

Features on different scales cause uneven convergence, slowing training or leading to suboptimal results.

Examples in Code (Using Scikit-learn)


- Min-Max Scaling:
```
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)
```
- Standardization:

```
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_standardized = scaler.fit_transform(X)
```

#Question 6: Compare Label Encoding and One-Hot Encoding. When would you prefer one over the other ? Hint: Consider categorical variables with ordinal vs. nominal relationships.

Ans - Both are techniques to convert categorical variables into numerical form, which is required by most machine learning algorithms.

##1. Label Encoding
Definition:

Assigns a unique integer to each category in a feature.

Example:
```
Color: [Red, Blue, Green] → [0, 1, 2]
```
Python Example:
```
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
encoded = le.fit_transform(['Red', 'Blue', 'Green', 'Red'])
# Output: [2, 0, 1, 2]
```

##When to Use:

- When the categorical variable is ordinal (i.e., has a meaningful order).

E.g., ["Low", "Medium", "High"] → [0, 1, 2]

- Useful when you want to reduce dimensionality.

###Drawbacks:

- For nominal (unordered) categories, label encoding introduces a false sense of order (e.g., 2 > 1), which can mislead algorithms like linear regression or SVM.

##2. One-Hot Encoding
Definition:

- Creates binary columns (0 or 1) for each category, indicating the presence of that category.

Example:
```
Color: [Red, Blue, Green] →
Red   Blue   Green  
 1      0      0  
 0      1      0  
 0      0      1
```
Python Example:
```
from sklearn.preprocessing import OneHotEncoder
import pandas as pd

df = pd.DataFrame({'Color': ['Red', 'Blue', 'Green']})
encoded = pd.get_dummies(df['Color'])
```
###When to Use:

- When the categorical variable is nominal (i.e., no meaningful order).

E.g., ["Male", "Female", "Other"], ["Red", "Green", "Blue"]

- Ensures equal weight for all categories.

###Drawbacks:

- Increases dimensionality (curse of dimensionality) — especially if the variable has many unique categories.

#Question 7: Google Play Store Dataset a). Analyze the relationship between app categories and ratings. Which categories have the
#highest/lowest average ratings, and what could be the possible reasons?
#Dataset: https://github.com/MasteriNeuron/datasets.git
#(Include your Python code and output in the code box below.)

Ans -

````python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset (assuming it's cloned or downloaded from the given GitHub repo)
url = "https://raw.githubusercontent.com/MasteriNeuron/datasets/main/googleplaystore.csv"
df = pd.read_csv(url)

# Display first few rows
print("Dataset preview:")
print(df.head(), "\n")

# Data cleaning
df = df.dropna(subset=['Rating', 'Category'])  # remove missing ratings or categories
df['Rating'] = pd.to_numeric(df['Rating'], errors='coerce')  # ensure numeric ratings
df = df.dropna(subset=['Rating'])  # drop any rows that failed conversion

# Compute average rating per category
category_rating = df.groupby('Category')['Rating'].mean().sort_values(ascending=False)

print("Average Rating by Category:\n")
print(category_rating)

# Plot top and bottom 10 categories by average rating
plt.figure(figsize=(12, 6))
sns.barplot(x=category_rating.head(10), y=category_rating.head(10).index, palette='viridis')
plt.title('Top 10 App Categories by Average Rating')
plt.xlabel('Average Rating')
plt.ylabel('Category')
plt.show()

plt.figure(figsize=(12, 6))
sns.barplot(x=category_rating.tail(10), y=category_rating.tail(10).index, palette='rocket')
plt.title('Bottom 10 App Categories by Average Rating')
plt.xlabel('Average Rating')
plt.ylabel('Category')
plt.show()
````

# Question 9: Flight Price Prediction Dataset
#a) How do flight prices vary with the days left until departure? Identify any exponential price
#surges and recommend the best booking window.
#b)Compare prices across airlines for the same route (e.g., Delhi-Mumbai). Which airlines are
#consistently cheaper/premium, and why?
#Dataset: https://github.com/MasteriNeuron/datasets.git
#(Include your Python code and output in the code box below.)

Ans-


````python
# Import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Load dataset
url = "https://raw.githubusercontent.com/MasteriNeuron/datasets/main/Flight_Price_Prediction.csv"
df = pd.read_csv(url)

# Display dataset preview
print("Dataset preview:")
print(df.head(), "\n")

# Basic cleaning
df = df.dropna(subset=['price', 'days_left'])
df['price'] = pd.to_numeric(df['price'], errors='coerce')
df['days_left'] = pd.to_numeric(df['days_left'], errors='coerce')
df = df.dropna(subset=['price', 'days_left'])

# ==============================
# a) Price vs. Days Left Analysis
# ==============================

# Group by days left and find average price
avg_price_days = df.groupby('days_left')['price'].mean().reset_index()

# Plot price vs. days left
plt.figure(figsize=(10,5))
sns.lineplot(data=avg_price_days, x='days_left', y='price', color='teal')
plt.title('Average Flight Price vs. Days Left Until Departure')
plt.xlabel('Days Left Until Departure')
plt.ylabel('Average Price (INR)')
plt.grid(True)
plt.show()

# Fit an exponential trendline for visualization
x = avg_price_days['days_left']
y = avg_price_days['price']
coeffs = np.polyfit(x, np.log(y), 1)
trend = np.exp(coeffs[1]) * np.exp(coeffs[0]*x)

plt.figure(figsize=(10,5))
plt.scatter(x, y, color='skyblue', label='Average Prices')
plt.plot(x, trend, color='red', label='Exponential Trend')
plt.title('Exponential Trend of Flight Prices vs. Days Left')
plt.xlabel('Days Left Until Departure')
plt.ylabel('Average Price (INR)')
plt.legend()
plt.grid(True)
plt.show()

# Identify potential exponential surge period (e.g., <7 days)
recent = avg_price_days[avg_price_days['days_left'] <= 7]
print("Average prices in the last 7 days before departure:\n")
print(recent)

# ==============================
# b) Airline Comparison for Same Route (Delhi-Mumbai)
# ==============================

# Filter for specific route
route_df = df[(df['source_city'] == 'Delhi') & (df['destination_city'] == 'Mumbai')]

airline_prices = route_df.groupby('airline')['price'].mean().sort_values()

print("\nAverage Prices for Delhi–Mumbai Route by Airline:\n")
print(airline_prices)

# Visualization
plt.figure(figsize=(12,6))
sns.barplot(x=airline_prices.values, y=airline_prices.index, palette='viridis')
plt.title('Average Flight Prices (Delhi–Mumbai) by Airline')
plt.xlabel('Average Price (INR)')
plt.ylabel('Airline')
plt.show()
````

#Question 10: HR Analytics Dataset
#a). What factors most strongly correlate with employee attrition? Use visualizations to show key drivers (e.g., satisfaction, overtime, salary).
#b). Are employees with more projects more likely to leave?
#Dataset: hr_analytics

Ans -

````python
# Import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset (assuming it's in the same repo or local folder)
url = "https://raw.githubusercontent.com/MasteriNeuron/datasets/main/hr_analytics.csv"
df = pd.read_csv(url)

# Display dataset preview
print("Dataset Preview:")
print(df.head(), "\n")

# Check basic info
print(df.info(), "\n")

# ====================================
# a) Correlation with Employee Attrition
# ====================================

# Encode Attrition column if it’s categorical
if df['Attrition'].dtype == 'object':
    df['Attrition'] = df['Attrition'].map({'Yes': 1, 'No': 0})

# Compute correlations with attrition
corr = df.corr(numeric_only=True)['Attrition'].sort_values(ascending=False)
print("Correlation of Features with Employee Attrition:\n")
print(corr, "\n")

# Select key features for visualization
key_features = ['JobSatisfaction', 'OverTime', 'MonthlyIncome', 'WorkLifeBalance', 'YearsAtCompany', 'Attrition']

# Convert OverTime to binary if not already numeric
if df['OverTime'].dtype == 'object':
    df['OverTime'] = df['OverTime'].map({'Yes': 1, 'No': 0})

# Visualize key factors
plt.figure(figsize=(10,6))
sns.heatmap(df[key_features].corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap: Key Drivers of Attrition')
plt.show()

# Visualize attrition by satisfaction level
plt.figure(figsize=(8,5))
sns.boxplot(x='Attrition', y='JobSatisfaction', data=df, palette='Set2')
plt.title('Job Satisfaction vs Attrition')
plt.xlabel('Attrition (1 = Left Company)')
plt.ylabel('Job Satisfaction')
plt.show()

# Visualize attrition by salary (Monthly Income)
plt.figure(figsize=(8,5))
sns.boxplot(x='Attrition', y='MonthlyIncome', data=df, palette='Set1')
plt.title('Monthly Income vs Attrition')
plt.xlabel('Attrition (1 = Left Company)')
plt.ylabel('Monthly Income')
plt.show()

# Visualize overtime effect
plt.figure(figsize=(6,4))
sns.countplot(x='OverTime', hue='Attrition', data=df, palette='magma')
plt.title('OverTime vs Attrition')
plt.xlabel('OverTime (1 = Yes)')
plt.ylabel('Number of Employees')
plt.show()

# ====================================
# b) Attrition vs Number of Projects
# ====================================

if 'NumProjects' in df.columns:
    project_col = 'NumProjects'
else:
    # Some datasets may call this 'NumberProject' or 'Projects'
    project_col = [col for col in df.columns if 'project' in col.lower()][0]

plt.figure(figsize=(8,5))
sns.boxplot(x='Attrition', y=project_col, data=df, palette='coolwarm')
plt.title(f'{project_col} vs Attrition')
plt.xlabel('Attrition (1 = Left Company)')
plt.ylabel('Number of Projects')
plt.show()

# Average projects by attrition
avg_projects = df.groupby('Attrition')[project_col].mean()
print(f"\nAverage Number of Projects by Attrition Status:\n{avg_projects}\n")
````
